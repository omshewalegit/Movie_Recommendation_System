```python
import ast 
ast.literal_eval
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj) :
        l.append(i['name'])
    return l    
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
dataset['genres'] = dataset['genres'].apply(convert)
```

```python
def convert3(obj):
    list = []
    count = 0
    for i in ast.literal_eval(obj):
        if count!=3:
            list.append(i['name'])
            count = count+1
        else:
            break    
    return list    
```

```python
dataset['cast'] = dataset['cast'].apply(convert3)
```

```python
def convert4(obj):
    list = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
    return list  
```

```python
dataset.rename(columns={'crew':'director'},inplace=True)
```

```python
dataset['director'] = dataset['director'].apply(convert4)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
df['tags'][0]
```

```python
stem(df['tags'][0])
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
def top10similar(movie):
    movie_idx = df[df['title']==movie].index[0]
    distances = similarity[movie_idx]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in movies_list:
        print(df.iloc[i[0]].title)
        
```
